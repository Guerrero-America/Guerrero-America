# **Gráfica de $f(x)=e^{(\frac{x}{2})}\cdot sen(\frac{x}{3})$**

# *Polinommio de McLaurin*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import sympy as sp
# definir x como simbolo
a=sp.Symbol('a')
# Funciones
f=sp.exp(a/2)*sp.sin(a/3)
print(f)
g=sp.diff(f,a) # primera derivada
print("f'=",g)
h=sp.diff(g,a) # segunda derivada
print("f''=",h)
i=sp.diff(h,a) # tercera derivada
print("f'''=",i)
```

```python
# Funcion y 3 poinomios de McLaurin
def f(x):
  return np.exp(x/2)*np.sin(x/3)
def g(x):
  return 1/3*x
def h(x):
  return (1/3*x)+(1/6*x**2)
def i(x):
  return (1/3*x)+(1/6*x**2)+(23/648*x**3)
#f=sp.exp(a/2)*sp.sin(a/3)
#g=sp.diff(f,a)
#print(f)
#print(g)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-100,100,1000)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, i(x))
# Característica<s de la gráfica
plt.title("Gráfica de $f(x)=e^{\\frac{x}{2}}\\cdot sen(\\frac{x}{3})$  y Polinomios de McLaurin")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-15, 30)
plt.ylim(-15, 60)
plt.xticks(np.arange(-15, 30, 5))
plt.yticks(np.arange(-15, 60, 5))
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend([r"$f(x)=e^{\frac{x}{2}}\cdot sen(\frac{x}{3})$",
            r"$g(x)=\frac{1}{3}x$",
            r"$h(x)=\frac{1}{3}x+\frac{1}{6}x^2$",
            r"$i(x)=\frac{1}{3}x+\frac{1}{6}x^2+\frac{23}{648}x^3$"]) #leyenda
plt.grid(color="grey")
plt.annotate("f(x)", (20,30),color="blue")
plt.annotate("g(x)", (25,10.5),color="orange")
plt.annotate("h(x)", (13,22),color="green")
plt.annotate("i(x)", (10,45),color="red")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
