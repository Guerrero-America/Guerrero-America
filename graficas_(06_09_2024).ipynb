# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x + 3**(-x)

def g(x):
    return x

def h(x):
    return -3**(-x)
```

```python
f(-5)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
plt.figure(figsize=(10,10))
# Graficar ambas funciones.
plt.plot(x, f(x), color="green")
plt.plot(x, g(x), color="orange")
#plt.legend({'g(x)= x'})
plt.legend({'h(x)= x','g(x)= -3**(-x)'})
#plt.plot(x, h(x))
#Caracteristicas de la grafica
plt.title("Gr치fica $f(x)=x-3**(-x)$")
plt.axhline(0, color="grey")
plt.axvline(0, color="grey")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

plt.xlim(-5, 5)
plt.ylim(-5, 5)
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
plt.grid(color="pink")
# Mostrarlo.
plt.show()
```
