# **Gr치fica de $f(x)=\sqrt{x+1}$**
# *Polinomios de Taylor*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    f=np.sqrt(x+1)
    return f
def g(x):
    return 1+(1/2)*x
def h(x):
    return 1+((1/2)*x)-((1/8)*x**2)
def i(x):
    return 1+((1/2)*x)-((1/8)*x**2)+((1/16)*x**3)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 20,1000)
```

```python
# Graficar funciones.
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, i(x))
#Caracter칤sticas de la gr치fica
plt.title("Gr치fica $f(x)=\sqrt{x+1}$  y Polinomios de Taylor") #Titulo
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-3, 7)
plt.ylim(-3, 7)
plt.xticks(np.arange(-3, 7, 1))
plt.yticks(np.arange(-3, 7, 1))
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend([r"$f(x)=\sqrt{x+1}$",
            r"$g(x)=1+\frac{1}{2}x$",
            r"$h(x)=1+\frac{1}{2}x-\frac{1}{8}x^2$",
            r"$i(x)=1+\frac{1}{2}x-\frac{1}{8}x^2+\frac{1}{16}x^3$"]) #leyenda
plt.grid(color="grey")
plt.annotate("f(x)", (4,1.7),color="blue")
plt.annotate("g(x)", (5,4),color="orange")
plt.annotate("h(x)", (5,0.7),color="green")
plt.annotate("i(x)", (4,6),color="red")
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
